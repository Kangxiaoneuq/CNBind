```python
import pandas as pd
import numpy as np
ori = pd.read_excel(r'TR60.xlsx')
name = ori.iloc[:1,0]
xl = ori.iloc[:1,1]###sequence
zimu = ori.iloc[:,5]###all_nucleotide_list

def Kmers_funct(seq, size): 
    return [seq[x:x+size] for x in range(len(seq) - size + 1)]


#计算该位置核苷酸到此出现的频率（密度）
dst=[]
for i in range(len(xl)):
    for j in range(len(xl[i])):
        a=xl[i][:j+1]#将a分成多段
        dst.append(a.count(a[j])/(j+1))###a[i]到此出现的次数
        
#该核苷酸在链上的总频率
fre = []
for i in range(len(xl)):
    one = Kmers_funct(xl[i],1)
    for j in range(len(one)):
        fre.append(one.count(one[j])/len(one))
#计算每一个字母的绝对位置
sl = []
for i in range(len(xl)):
    one = Kmers_funct(xl[i],1)
    for j in range(len(one)):
        sl.append((j+1)/len(one))

nps = []
for i in range(len(xl)):
    if len(xl[i])%2 == 1:
        for j in range(int((len(xl[i])+1)/2),0,-1):
            nps.append(j/((len(xl[i])+1)/2))
        for j in range(2,int((len(xl[i])+3)/2)):
            nps.append(j/((len(xl[i])+1)/2))
    else:
        for j in range(int(len(xl[i])/2),0,-1):
            nps.append(j/(len(xl[i])/2))
        for j in range(1,int((len(xl[i])+2)/2)):
            nps.append(j/(len(xl[i])/2))
            
```

```python
#NDC+NDS
from scipy.spatial.distance import pdist
import copy
import os
import networkx as nx

ori = pd.read_excel(r'TR60.xlsx')
Name = ori.iloc[:1,0]#sequence_name
xl = ori.iloc[:1,1]#sequence

name = []
for i in range(len(Name)):
    name.append(Name[i])
```

```python
path = "D:\\jupyter notebook files\\xdp2\\PDB\\"######PDB

df_empty = pd.DataFrame()
NDS = []
NDC = []
COSS = []
COSC = []
CHEBS = []
CHEBC = []
for n in range(len(name)):    
    x1 = []
    y1 = []
    z1 = []
    for line in open(path + name[n] +'.pdb'):
        list = line.split()
        if list[0] == 'ATOM'or list[0] == 'HETATM':
            if list[2] == "C1'":
                x1.append(float(list[6]))
                y1.append(float(list[7]))
                z1.append(float(list[8]))
    zhixin = np.array([sum(x1)/len(x1),sum(y1)/len(y1),sum(z1)/len(z1)])
    
    for i in range(len(x1)):
        nds = []
        coss = []
        chebs = []
        for j in range(len(x1)):
            if i!=j:
                a = np.array([x1[i],y1[i],z1[i]])
                b = np.array([x1[j],y1[j],z1[j]])
                X = np.vstack([a,b])
                c = pdist(X)
                nds.append(c[0])
                coss.append(np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b)))
                chebs.append(np.abs(a-b).max())
        NDS.append(sum(nds)/len(x1))
        COSS.append(sum(coss)/len(x1))
        CHEBS.append(sum(chebs)/len(x1))
        d = np.array([x1[i],y1[i],z1[i]])
        Y = np.vstack([d,zhixin])
        e = pdist(Y)
        NDC.append(e[0])
        COSC.append(np.dot(d,zhixin)/(np.linalg.norm(d)*np.linalg.norm(zhixin)))
        CHEBC.append(np.abs(d-zhixin).max())
zong = {'NDS':NDS,
        'NDC':NDC,
        'COSS':COSS,
        'COSC':COSC,
        'CHEBS':CHEBS,
        'CHEBC':CHEBC,}
```
